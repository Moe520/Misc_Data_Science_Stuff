```python
data = pd.read_csv("hgsc.csv")
```

```python
data['class'] = data['class'].astype('category')
```

```python
data['class'] = data['class'].cat.codes
```

```python
data.head()
```

```python
X = data.drop(["class"],axis=1)
Y = data["class"] 
```

```python
#Description of begging function from sklearn manual
#A Bagging classifier is an ensemble meta-estimator that fits base classifiers each on random subsets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction.
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
f = QuadraticDiscriminantAnalysis()

from sklearn.ensemble import BaggingClassifier
bagger = BaggingClassifier(base_estimator=f,n_estimators=100,max_samples=100)
bagger.fit(X,Y)
QDApreds = bagger.predict(X)

sum(QDApreds==Y)/len(Y)
```

```python

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
f = LinearDiscriminantAnalysis()

from sklearn.ensemble import BaggingClassifier
bagger = BaggingClassifier(base_estimator=f,n_estimators=100,max_samples=100)
bagger.fit(X,Y)
LDApreds = bagger.predict(X)

sum(LDApreds==Y)/len(Y)
```

```python
from sklearn import svm
f= svm.SVC()

from sklearn.ensemble import BaggingClassifier
bagger = BaggingClassifier(base_estimator=f,n_estimators=100,max_samples=100)
bagger.fit(X,Y)
SVMpreds = bagger.predict(X)

sum(SVMpreds==Y)/len(Y)
```

```python
from sklearn.ensemble import RandomForestClassifier
f = RandomForestClassifier()

from sklearn.ensemble import BaggingClassifier
bagger = BaggingClassifier(base_estimator=f,n_estimators=100,max_samples=100)
bagger.fit(X,Y)
Forestpreds = bagger.predict(X)

sum(Forestpreds==Y)/len(Y)
```

```python
frames = [LDApreds,QDApreds,SVMpreds,Forestpreds]
result = pd.DataFrame(frames)
result=result.transpose()
```

```python
result.columns = ['LDA','QDA','SVM','Random Forest']
```
